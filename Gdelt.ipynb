```python
import gdelt
import pandas as pd
```

```python
gd2 = gdelt.gdelt(version=2)


events_df = gd2.Search(
    ["2015 Feb 18", "2025 Dec 31"],
    table="events",
    coverage=False,
    output="pd"
)

events_df['SQLDATE'] = pd.to_datetime(
    events_df['SQLDATE'].astype(str),
    format='%Y%m%d',
    errors='coerce'
)

events_df['DATEADDED'] = pd.to_datetime(
    events_df['DATEADDED'].astype(str),
    format='%Y%m%d%H%M%S',
    errors='coerce'
)
```

```python
events_df['event_lag_days'] = (
    events_df['DATEADDED'] - events_df['SQLDATE']
).dt.days

events_recent = events_df[
    (events_df['event_lag_days'] >= 0) &
    (events_df['event_lag_days'] <= 2)
]
```

```python
events_recent['SQLDATE'] = pd.to_datetime(events_recent['SQLDATE'])

# Create Month column (first day of month)
events_recent['Month'] = events_recent['SQLDATE'].values.astype('datetime64[M]')

# Create ISO week (year-week format)
# Week number in the month (1 = first week, etc.)
events_recent['WeekOfMonth'] = events_recent['SQLDATE'].apply(
    lambda d: ((d.day - 1) // 7) + 1
)
```

```python
events_recent
```

```python
eu_countries = [
    "AUT","BEL","FIN","FRA","DEU",
    "IRL","ITA","LUX","NLD","PRT","ESP"
]

actor1 = events_recent[
    events_recent['Actor1CountryCode'].isin(eu_countries)
].copy()

actor1['country'] = actor1['Actor1CountryCode']
actor1['role'] = 'actor1'

actor2 = events_recent[
    events_recent['Actor2CountryCode'].isin(eu_countries)
].copy()

actor2['country'] = actor2['Actor2CountryCode']
actor2['role'] = 'actor2'

long_df = pd.concat([actor1, actor2], ignore_index=True)
```

```python
event_batches = {
    # Actual economic actions (flows, sanctions, aid)
    'direct_economic_action': [
        '061',   # Cooperate economically
        '071',   # Provide economic aid
        '085',   # Ease economic sanctions
        '1621',  # Reduce or stop economic assistance
        '163'    # Impose sanctions
    ],

    # Policy decisions / institutional change (not signals)
    'economic_policy_action': [
        '1042',  # Demand policy change
        '081'    # Ease administrative sanctions
    ],

    # Forward-looking signals / expectations
    'economic_policy_signal': [
        '014',   # Consider policy option
        '0211',  # Appeal for economic cooperation
        '0231',  # Appeal for economic aid
        '0311',  # Express intent to cooperate economically
        '0331',  # Express intent to provide economic aid
        '0354'   # Express intent to ease economic sanctions
    ],

    # Pressure via threats or demands
    'economic_threat_pressure': [
        '1011',  # Demand economic cooperation
        '1031',  # Demand economic aid
        '1054',  # Demand easing of economic sanctions
        '1311',  # Threaten to reduce or stop aid
        '1312'   # Threaten with sanctions
    ],

    # Explicit rejection / refusal
    'economic_rejection': [
        '1211',  # Reject economic cooperation
        '1221',  # Reject request for economic aid
        '1244'   # Refuse to ease economic sanctions
    ]
}


def map_batch(code):
    for k, v in event_batches.items():
        if str(code) in v:
            return k
    return None

long_df['event_batch'] = long_df['EventCode'].astype(str).apply(map_batch)

long_df = long_df[long_df['event_batch'].notna()]
```

```python
long_df['EventCode'].value_counts().head(20)
```

```python
monthly_agg = (
    long_df
    .groupby([
        'Month',
        'country',
        'role',
        'event_batch'
    ])
    .agg(
        goldstein_mean=('GoldsteinScale', 'mean'),
        goldstein_median=('GoldsteinScale', 'median'),
        event_count=('GoldsteinScale', 'size')
    )
    .reset_index()
)
```

```python
monthly_agg['goldstein_weighted'] = (
    monthly_agg['goldstein_mean'] * monthly_agg['event_count']
)

monthly_country_role = (
    monthly_agg
    .groupby(['Month', 'country', 'role'])
    .agg(
        monthly_goldstein=('goldstein_weighted', lambda x: x.sum() / monthly_agg.loc[x.index, 'event_count'].sum()),
        monthly_event_count=('event_count', 'sum')
    )
    .reset_index()
)

pivot = monthly_country_role.pivot(
    index=['Month', 'country'],
    columns='role',
    values='monthly_goldstein'
).reset_index()

pivot['net_goldstein'] = pivot.get('actor1', 0) - pivot.get('actor2', 0)
```
